```python
from sortedcontainers import SortedList
import datetime

class CustomItem:
    def __init__(self, time_str, val1, val2):
        self.time_str = time_str
        self.val1 = val1
        self.val2 = val2

    def __repr__(self):
        return f'CustomItem({self.time_str}, {self.val1}, {self.val2})'

    def __lt__(self, other):
        return self.time_str < other.time_str or self.val1 < other.val1


# 创建一个包含列表的 SortedList
sorted_list = SortedList([
    CustomItem("9:15:00", 1, 4),
    CustomItem("9:00:00", 6, 9),
    CustomItem("9:16:00", 6, 5),
    CustomItem("9:16:00", 3, 5)
])

# # 输出排序后的列表
for item in sorted_list:
    print(item)

sorted_list.add(CustomItem("9:16:00", 4, 5))
for item in sorted_list:
    print(item.__repr__())
```

```python
# from math import max
def total_intersection_depth(segments):
    events = []
    for start, end in segments:
        events.append((start, 1))
        events.append((end, -1))
    events.sort()

    total_depth = 0
    depth = 0
    covered = set()
    prev_x = None
    for x, delta in events:
        if prev_x is not None and covered:
            total_depth += depth * (x - prev_x)
        if delta == 1:
            cnt = max(covered) if len(covered) > 0 else 0
            for y in range(x, cnt + 1):
                covered.add(y)
        else:
            covered.remove(x)
        depth += delta
        prev_x = x

    return total_depth

# Example usage
segments = [(1, 3), (2, 5), (4, 7), (6, 9)]
print(total_intersection_depth(segments))
```
